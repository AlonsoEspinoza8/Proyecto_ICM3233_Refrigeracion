```python
T_ref_inviero = 18
T_ref_verano = 21 

def GH_calculo(temp_totales_max_verano, temp_totales_min_verano, temp_totales_max_invierno, temp_totales_min_invierno):
    #Verano
    GH_verano = 0
    contador = 0
    temporal = []
    for t in range(len(temp_totales_max_verano)):
        contador += 1
        t_media = (temp_totales_max_verano[t] + temp_totales_min_verano[t])*0.5



        if t_media - T_ref_verano  > 0:
            temporal.append(t_media)
        
        if contador == 60:
            if temporal != []:
                media = mean(temporal)
                GH_verano += media - T_ref_verano
            contador = 0
            temporal = []

    print(f'GH verano: {GH_verano:.2f}')

    #Invierno
    GH_invierno = 0
    contador = 0
    temporal = []
    for t in range(len(temp_totales_min_invierno)):
        contador += 1
        t_media = (temp_totales_max_invierno[t] + temp_totales_min_invierno[t])*0.5

        if T_ref_inviero - t_media  > 0:
            temporal.append(t_media)
        
        if contador == 60:
            if temporal != []:
                media = mean(temporal)
                GH_invierno += T_ref_inviero - media
            contador = 0
            temporal = []

    print(f'GH invierno: {GH_invierno:.2f}')
    return [GH_verano, GH_invierno]
```

```python
GHV23, GHI23 = GH_calculo(temp_totales_max_verano_23, temp_totales_min_verano_23, temp_totales_max_invierno_23, temp_totales_min_invierno_23)
GHV24, GHI24 = GH_calculo(temp_totales_max_verano_24, temp_totales_min_verano_24, temp_totales_max_invierno_24, temp_totales_min_invierno_24)
```
