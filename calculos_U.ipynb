# Tabiques

## Tabique tipo 1

- Dos planchas Yeso Cartón ST de 15 mm
- Aislación Ainglass R-141 60 mm
- 2 capas de banda acústica Fonodan

```python
h = 2715e-3 # m. Altura del segundo piso
e_yeso = 15e-3 # m
e_aislante = 60e-3 # m
k_yeso_st = 0.24 # W/mK https://volcan.cl/wp-content/uploads/2022/03/planchas_de_yeso_carton_para_uso_interior_ces.pdf
k_aislante = 0.04 # W/mK 
k_banda_acustica = 0.04 # W/mK
R_yeso_st = e_yeso / (k_yeso_st)
R_aislante = e_aislante / (k_aislante)
R_banda_acustica = e_aislante / (k_banda_acustica * (h - 0.008)) # Restando el espesor de la banda acústica
R_T1 = 2* R_yeso_st + R_aislante
U_T1 = 1 / R_T1
print(f'Coeficiente global de transferencia de calor tabique tipo 1: {U_T1:.2f} W/Km^2')
```

# Ventanas

## Ventanas V01-V02

- Cristal incoloro 6 mm
- Cámara de aire 12 mm
- Cristal laminado blindex 6 mm

Tomamos las conductividades térmicas del aire quieto y del vidrio del anexo A1 de la norma chilena NCh853

```python
k_aire_quieto = (0.031 - 0.024) / (100) * (25) + 0.024  # W/mK, interpolación lineal entre 0 y 100 C (NCh853 Anexo A1)
print(f'Conductividad térmica aire quieto a 25 C: {k_aire_quieto:.4f} W/mK')
e_aire = 12e-3 # mm
k_vidrio = 1.2 # W/mK, NCh853 Anexo A1
e_vidrio = 6e-3 # m
R_aire = e_aire / (k_aire_quieto)
R_vidrio = e_vidrio / (k_vidrio)
R_V01 = R_aire + 2 * R_vidrio
U_V01 = 1 / R_V01
print(f'Coeficiente global de transferencia de calor ventana tipo 1: {U_V01:.2f} W/Km^2')
U_V02 = U_V01  # Mismo cálculo que ventana tipo 1
```

## Ventana MV17
